```python
from skimage import io
from skimage.color import rgb2gray

import numpy as np

import pywt

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

# Import Sample Image

```python
image = io.imread("./Feynman.png")
print(image.shape)
image = rgb2gray(image)
print(image.shape)
```

```python
fig = plt.figure(figsize=(8., 12.))
gs = gridspec.GridSpec(1, 1)
gs.update(wspace=0.025, hspace=0.0)
#
ax = plt.subplot(gs[0])
ax.imshow(image, cmap="gray")
ax.axis("off")
#
plt.show()
```

<br><br><br>

# Wavelet Transformed Image:

```python
LL, (LH, HL, HH) = pywt.dwt2(data=image, wavelet="haar", axes=(0, 1))
LLL, (LHL, HLL, HHL) = pywt.dwt2(data=LL, wavelet="haar", axes=(0, 1))
```

```python
fig = plt.figure(figsize=(13., 18.25))
gs = gridspec.GridSpec(2, 2)
gs.update(wspace=0.0125, hspace=0.0125)

gs2 = gridspec.GridSpecFromSubplotSpec(
    2, 2, subplot_spec=gs[0, 0], wspace=0.00625, hspace=0.00625)
#
ax = plt.subplot(gs2[0, 0])
ax.imshow(LLL, cmap="gray")
ax.axis("off")
#
ax = plt.subplot(gs2[0, 1])
ax.imshow(LHL, cmap="gray")
ax.axis("off")
#
ax = plt.subplot(gs2[1, 0])
ax.imshow(HLL, cmap="gray")
ax.axis("off")
#
ax = plt.subplot(gs2[1, 1])
ax.imshow(HHL, cmap="gray")
ax.axis("off")
#
ax = plt.subplot(gs[0, 1])
ax.imshow(LH, cmap="gray")
ax.axis("off")
#
ax = plt.subplot(gs[1, 0])
ax.imshow(HL, cmap="gray")
ax.axis("off")
#
ax = plt.subplot(gs[1, 1])
ax.imshow(HH, cmap="gray")
ax.axis("off")
#
plt.show()
```
